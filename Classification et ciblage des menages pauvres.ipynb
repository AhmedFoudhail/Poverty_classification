+ <div style = "font-family: 'Georgia' ; font-size: 12pt ; line-height: 1.75; text-align: justify" > <b> Solution proposée : </b> </div>

<div style = "font-family: 'Cambria Math' ; font-size: 12pt ; line-height: 1.5; text-align: justify" > Ce projet de machine learning vise à développer un modèle de classification robuste utilisant des techniques de régression logistique pour prédire le statut de pauvreté des ménages en se basant sur les données de l'EPCVM. En appliquant les étapes classiques d'un projet de machine learning, incluant la préparation des données, la sélection des caractéristiques, la construction du modèle, l'évaluation et l'interprétation des résultats, nous chercherons à créer un outil fiable pour identifier les ménages en situation de pauvreté.
 </div> 



+ <div style = "font-family: 'Georgia' ; font-size: 12pt ; line-height: 1.5; text-align: justify" > <b> Méthodologie : </b> </div>


<div style = "font-family: 'Cambria Math' ; font-size: 12pt ; line-height: 1.5; text-align: justify" > La méthodologie principale de ce projet reposera sur l'utilisation d'une régression logistique, une méthode adaptée pour la classification binaire. À partir de cette approche, nous construirons un modèle statistique pour prédire avec précision si un ménage est pauvre ou non pauvre en se basant sur les variables disponibles dans les données de l'EPCVM.
 </div> 
 
 
<div style = "font-family: 'Cambria Math' ; font-size: 12pt ; line-height: 1.5; text-align: justify" > La fonction logistique est définie comme : <br><br>

<div style="text-align:center; font-size:150%;">
\[ P(Y=1|X) = \frac{1}{1 + e^{-(\beta_0 + \beta_1X_1 + \beta_2X_2 + ... + \beta_nX_n)}} \]
</div>

Où :
- $  P(Y=1|X)  $ est la probabilité conditionnelle que la variable dépendante $Y$ soit égale à 1 pour un ensemble de variables indépendantes $X$.
- $ \beta_0, \beta_1, \beta_2, ..., \beta_n $ sont les coefficients du modèle.
- $ X_1, X_2, ..., X_n $ sont les variables indépendantes.

 </div>

```python
from scipy.stats import chi2_contingency
# Sélectionner les variables catégorielles
cat_vars = ['milieu', 'stat_matCM', 'sexeCM', 'wilaya',  'sait_lire_ecrire',
             'Diplome_eleve', 'assurence_maladie', 'type_habitat', 'source_eau', 'energie_cuisine', 
            'source_eclairage', 'type_emploi', 'taille_classes', 'age_classe']
# Initialiser les listes pour stocker les résultats
var_names = []
chi2_stats = []
p_values = []
cramer_vs = []

# Parcourir toutes les variables catégorielles
for var in cat_vars:
    # Calculer le tableau de contingence
    contingency_table = pd.crosstab(mymen['pauvre'], mymen[var])
    # Calculer la statistique de test du Chi-deux et la p-valeur
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    # Calculer le coefficient V de Cramer
    n = contingency_table.sum().sum()
    phi2 = chi2/n
    r,k = contingency_table.shape
    phi2corr = max(0, phi2-((k-1)*(r-1))/(n-1))
    rc = r-((r-1)**2)/(n-1)
    kc = k-((k-1)**2)/(n-1)
    cramer_v = np.sqrt(phi2corr/min(rc-1,kc-1))
    # Ajouter les résultats aux listes correspondantes
    var_names.append(var)
    chi2_stats.append(chi2)
    p_values.append(p)
    cramer_vs.append(cramer_v)

# Créer un DataFrame avec les résultats
results_df = pd.DataFrame({
    'Variable': var_names,
    'Chi2': chi2_stats,
    'P-valeur': p_values,
    'V de Cramer': cramer_vs
})

# Trier le DataFrame par ordre croissant de V de Cramer
results_df.sort_values(by='V de Cramer', inplace=True,ascending=False)

# Afficher le tableau des résultats
print(results_df)
```

```python
def plot_cm(y_true, y_pred, title, figsize=(5,4)):
    cm = confusion_matrix(y_true, y_pred, labels=np.unique(y_true))
    cm_sum = np.sum(cm, axis=1, keepdims=True)
    cm_perc = cm / cm_sum.astype(float) * 100
    annot = np.empty_like(cm).astype(str)
    nrows, ncols = cm.shape
    for i in range(nrows):
        for j in range(ncols):
            c = cm[i, j]
            p = cm_perc[i, j]
            if i == j:
                s = cm_sum[i]
                annot[i, j] = '%.1f%%\n%d/%d' % (p, c, s)
            elif c == 0:
                annot[i, j] = ''
            else:
                annot[i, j] = '%.1f%%\n%d' % (p, c)
    cm = pd.DataFrame(cm, index=np.unique(y_true), columns=np.unique(y_true))
    cm.index.name = 'Actual'
    cm.columns.name = 'Predicted'
    fig, ax = plt.subplots(figsize=figsize)
    plt.title(title)
    sns.heatmap(cm, cmap= "YlGnBu", annot=annot, fmt='', ax=ax)
```
